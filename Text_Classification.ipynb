```python
#CREATING FIT FUNCTION FROM SCRATCH
def fit(x, y):
    result = {}
    #storing total training data points
    result['total_data'] = len(y)
    #gives all distinct outputs possible
    class_values = set(y)
    for current_class in class_values:
        result[current_class] = {}
        #getting a boolean array where value will be true if the condition is true
        current_class_rows = (y == current_class)
        #getting all input with class output as current class
        x_current = x[current_class_rows]
        #getting all output with class output as current class
        y_current = y[current_class_rows]
        #getting number of features
        num_features = x.shape[1]
        #total count of current class
        result[current_class]["total_count"] = len(y_current)
        #storing total frequency of each word for current class
        for i in range(num_features):
            result[current_class][i] = x_current[:, i].sum()
        
        result[current_class]['total_words'] = sum(result[current_class].values())
    
    return result
```

```python
#to calculate prob of an input corresponding to each output class
def probability(dictionary, x, current_class):
    output = np.log(dictionary[current_class]["total_count"]) - np.log(dictionary["total_data"])# probability P(y == a)
    num_features = len(dictionary[current_class].keys()) - 2 #subtracting 2 because we have added key of total count and total words
    for i in range(num_features):
        num_word_i = dictionary[current_class][i] + 1 #+1 of laplace correction
        total_words = dictionary[current_class]["total_words"] + num_features #num features of laplace correction
        prob_i = x[i] * (np.log(num_word_i) - np.log(total_words))# probability P(xi == xi/ y == a)
        output += prob_i
    
    return output
```

```python
#to calculate output of a single data point
def predictSinglePoint(dictionary, x):
    #calculate probability p(x = / y = ai)*p(ai) for each class
    classes = dictionary.keys()
    output_class = -1
    best_p = -100000
    first_run = True
    for current_class in classes:
        if current_class == 'total_data': continue
        p_current_class = probability(dictionary, x, current_class)
        if (p_current_class > best_p or first_run):
            output_class = current_class
            best_p = p_current_class
        first_run = False
    
    return output_class
```

```python
#function to predict output for a given set of input
def predict(dictionary, xtest):
    y_pred = []
    for x in xtest:
        #find the class with highest probability corresponding to input x
        x_class = predictSinglePoint(dictionary, x)
        y_pred.append(x_class)
    
    return y_pred
```

```python
#USING SELF IMPLEMENTED MULTINOMIAL NAIVE BAYES
dictionary = fit(xtrain,ytrain)
dictionary.keys()
```

```python
#testing multinomial naive bayes(self implemented)
ypredict_self = predict(dictionary, xtest)
```
